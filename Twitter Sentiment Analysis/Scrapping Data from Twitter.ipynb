```python
hashtag_tweets = tweepy.Cursor(api.search, q="#Covid19", tweet_mode='extended', lang="en", wait_on_rate_limit=True).items(50000)
tweets = []
tweet_point = []
for tweet in hashtag_tweets:
    username = tweet.author.screen_name
    text = tweet.full_text
    tweet_point = [username , text]
    tweets.append(tweet_point)
```
